### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_tienda = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()
```

```python
print(f'Ingreso total tienda 1: ${ingreso_tienda:,.0f}')
print(f'Ingreso total tienda 2: ${ingreso_tienda2:,.0f}')
print(f'Ingreso total tienda 3: ${ingreso_tienda3:,.0f}')
print(f'Ingreso total tienda 4: ${ingreso_tienda4:,.0f}')
```

# 2. Ventas por categoría

```python
ventas_categoria_t1 = tienda['Categoría del Producto'].value_counts()
ventas_categoria_t2 = tienda2['Categoría del Producto'].value_counts()
ventas_categoria_t3 = tienda3['Categoría del Producto'].value_counts()
ventas_categoria_t4 = tienda4['Categoría del Producto'].value_counts()
```

```python
print("Ventas por categoría - Tienda 1")
print(ventas_categoria_t1, '\n')

print("Ventas por categoría - Tienda 2")
print(ventas_categoria_t2, '\n')

print("Ventas por categoría - Tienda 3")
print(ventas_categoria_t3, '\n')

print("Ventas por categoría - Tienda 4")
print(ventas_categoria_t4)
```

# 3. Calificación promedio de la tienda


```python
calificacion_t1 = tienda['Calificación'].mean()
calificacion_t2 = tienda2['Calificación'].mean()
calificacion_t3 = tienda3['Calificación'].mean()
calificacion_t4 = tienda4['Calificación'].mean()
```

```python
print(f'Calificación promedio Tienda 1: {calificacion_t1:.2f}')
print(f'Calificación promedio Tienda 2: {calificacion_t2:.2f}')
print(f'Calificación promedio Tienda 3: {calificacion_t3:.2f}')
print(f'Calificación promedio Tienda 4: {calificacion_t4:.2f}')
```

# 4. Productos más y menos vendidos

```python
productos_vendididos_t1 = tienda['Producto'].value_counts()
productos_vendididos_t2 = tienda2['Producto'].value_counts()
productos_vendididos_t3 = tienda3['Producto'].value_counts()
productos_vendididos_t4 = tienda4['Producto'].value_counts()
```

```python
print('Tienda 1')
print('Producto más vendido:')
print(productos_vendididos_t1.head(1))
print('Productos menos vendidos:')
print(productos_vendididos_t1.tail(1))
print('\n')

print('Tienda 2')
print('Producto más vendido:')
print(productos_vendididos_t2.head(1))
print('Productos menos vendidos:')
print(productos_vendididos_t2.tail(1))
print('\n')

print('Tienda 3')
print('Producto más vendido:')
print(productos_vendididos_t3.head(1))
print('Productos menos vendidos:')
print(productos_vendididos_t3.tail(1))
print('\n')

print('Tienda 4')
print('Producto más vendido:')
print(productos_vendididos_t4.head(1))
print('Productos menos vendidos:')
print(productos_vendididos_t4.tail(1))
```

# 5. Envío promedio por tienda

```python
envio_prom_t1 = tienda['Costo de envío'].mean()
envio_prom_t2 = tienda2['Costo de envío'].mean()
envio_prom_t3 = tienda3['Costo de envío'].mean()
envio_prom_t4 = tienda4['Costo de envío'].mean()
```

```python
print(f'Envío promedio Tienda 1: {envio_prom_t1:.2f}')
print(f'Envío promedio Tienda 2: {envio_prom_t2:.2f}')
print(f'Envío promedio Tienda 3: {envio_prom_t3:.2f}')
print(f'Envío promedio Tienda 4: {envio_prom_t4:.2f}')
```

# Resumen en Gráficos

```python
resumen = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Ingreso Total': [ingreso_tienda, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4],
    'Calificación Promedio': [calificacion_t1, calificacion_t2, calificacion_t3, calificacion_t4],
    'Envío Promedio': [envio_prom_t1, envio_prom_t2, envio_prom_t3, envio_prom_t4]
})
```

## - Ingreso total por tienda

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.bar(resumen['Tienda'], resumen['Ingreso Total'])
plt.title('Ingreso total por tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingreso total')
plt.show()
```

## - Calificación promedio por tienda

```python
plt.figure(figsize=(8,5))
plt.bar(resumen['Tienda'], resumen['Calificación Promedio'])
plt.title('Calificación promedio por tienda')
plt.xlabel('Tienda')
plt.ylabel('Calificación promedio')
plt.ylim(0,5)
plt.show()
```

## - Envío promedio por tienda

```python
plt.figure(figsize=(8,5))
plt.bar(resumen['Tienda'], resumen['Envío Promedio'])
plt.title('Costo de envío promedio por tienda')
plt.xlabel('Tienda')
plt.ylabel('Costo de envío')
plt.show()
```
